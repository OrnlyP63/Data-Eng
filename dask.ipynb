```python
import numpy as np
import dask
```

```python
@dask.delayed
class k:
    def __init__(self, a, b):
        self.sum = a + b
        self.mul = a*b
    
    def div(self, k):
        return k * (self.sum - self.mul)
```

```python
%%time
output = []
data = [1, 2, 3, 4, 5]
for x in data:
    a = k(x, x*2)
    output.append(a.div(2))
    
total = dask.delayed(sum)(output)
```

```python
total.compute()
```

```python
total.visualize()
```
